```python
# Convert the image into RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
plt.imshow(img[...,::-1]) #this conversion can be performed inline!
#plt.show()
```

```python
# Convert the image into gray scale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap = 'gray')
```

```python
# Plot the three channels of the image
fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize = (20, 20))
for i in range(0, 3):
    ax = axs[i]
    ax.imshow(img_rgb[:, :, i], cmap = 'gray')
plt.show()
```

```python
# Transform the image into HSV and HLS models
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
# Plot the converted images
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 20))
ax1.imshow(img_hsv)
ax2.imshow(img_hls)
plt.show()
```

```python
# Copy the image
img_copy = img.copy()
img_copy = cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB)
# Draw a rectangle 
cv2.rectangle(img_copy, pt1 = (50, 50), pt2 = (200, 200), 
              color = (255, 0, 0), thickness = 5)
plt.imshow(img_copy)
```

```python
#circle
cv2.circle(img_copy, center = (125, 125), radius = 50, 
           color = (0, 0, 255), thickness = 5)
plt.imshow(img_copy)
```

```python
# Add text 
cv2.putText(img_copy, text = "Lenna", 
            org = (50, 50),
            fontFace = cv2.FONT_HERSHEY_DUPLEX, 
            fontScale = 1, 
            color = (0, 255, 0), 
            thickness = 2, 
            lineType = cv2.LINE_AA)
plt.imshow(img_copy)
```
